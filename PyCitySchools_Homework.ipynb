```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Caluclate School Count
unique_schools = school_data_complete["school_name"].unique()
school_count = len(unique_schools)

school_count
```

```python
#Calculate Total Students
total_students = school_data_complete["student_name"].count()
total_students
```

```python
#Calculate Total Budget
total_budget = school_data["budget"].sum()
total_budget
```

```python
#Calculate Average Math Score
average_math_score = school_data_complete["math_score"].mean()
average_math_score
```

```python
#Calculate Average Reading Score
average_reading_score = school_data_complete["reading_score"].mean()
average_reading_score
```

```python
#Calculate Passing Math %
total_math_pass_score = school_data_complete.loc[school_data_complete["math_score"] >= 70]["student_name"].count()

percent_math_pass = total_math_pass_score/total_students
percent_math_pass
```

```python
#Calculate Passing Reading %
total_reading_pass_score = school_data_complete.loc[school_data_complete["reading_score"] >= 70]["student_name"].count()

percent_reading_pass = total_reading_pass_score/total_students
percent_reading_pass
```

```python
#Calculate Overall Passing % (both math and reading)
combo_student_pass = school_data_complete[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)]['student_name'].count()
combo_passing_percentage = combo_student_pass/total_students
combo_passing_percentage
```

```python
#Enter into dataframe
district_summary_df = pd.DataFrame({
    
    "Total Schools": [school_count],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math_score],
    "Average Reading Score": [average_reading_score],
    "% Passing Math": [percent_math_pass],
    "% Passing Reading":[percent_reading_pass],
    "Overall Passing Rate": [combo_passing_percentage]
})
    
district_summary_df
```

```python
#Formatted Data
district_summary_df.style.format ({"Total Budget": "${:,.0f}", 
                       "Average Reading Score": "{:.1f}", 
                       "Average Math Score": "{:.1f}", 
                       "% Passing Math": "{:.1%}", 
                       "% Passing Reading": "{:.1%}", 
                       "Overall Passing Rate": "{:.1%}"})
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#School Summary Info

#Group by School Name
sort_school_df = school_data_complete.groupby(['school_name'])
sort_school_df.count()

#Configure School Type
school_types = school_data.set_index(["school_name"])["type"]
school_types

#Total Students by School
students_per_school = school_data_complete.groupby("school_name").count()["student_name"]
students_per_school

#Budget Per School
budget_per_school = school_data_complete.groupby("school_name").mean()["budget"]
budget_per_school

#Budget per Student
per_student_budget = budget_per_school/students_per_school
per_student_budget

#Average Math Score by School
avg_math_school = school_data_complete.groupby("school_name").mean()["math_score"]
avg_math_school

#Average Reading Score
avg_reading_school = school_data_complete.groupby("school_name").mean()["reading_score"]
avg_reading_school

#% Passing Math
pass_math_pct = school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name").count()["Student ID"]/students_per_school 
pass_math_pct

# #% Passing Reading
pass_reading_pct = school_data_complete[school_data_complete["reading_score"] >= 70].groupby("school_name").count()["Student ID"]/students_per_school 
pass_reading_pct

# #% Passing Math & Reading
pass_combo_df = school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)]

pass_combo_df
pass_combo_pct = pass_combo_df.groupby("school_name").count()["student_name"]/students_per_school
pass_combo_pct
```

```python
# School Summary Dataframe

school_summary_df = pd.DataFrame({
    
                                "School Type": school_types,
                               "Total Students": students_per_school,
                               "Total School Budget": budget_per_school,
                              "Per Student Budget": per_student_budget,
                              "Average Math Score": avg_math_school,
                              "Average Reading Score": avg_reading_school,
                              "% Passing Math": pass_math_pct,
                              "% Passing Reading": pass_read_pct,
                              "% Overall Passing": pass_combo_pct
                            })

school_summary_df
```

```python
#Formatted Data
school_summary_df.style.format ({"Total School Budget": "${:,.0f}",
                                 "Per Student Budget": "${:,.0f}",
                       "Average Reading Score": "{:.1f}", 
                       "Average Math Score": "{:.1f}", 
                       "% Passing Math": "{:.1%}", 
                       "% Passing Reading": "{:.1%}", 
                       "% Overall Passing": "{:.1%}"})
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Define bins and group names

bins = [0, 584, 629, 644, 675]

group_names = ["$0-$584", "$585-$629", "$630-$644", "$645-$675"]
```

```python
school_summary_df["Spend per Student"] = pd.cut(school_summary_df["Per Student Budget"], bins, labels=group_names)

school_spend = school_summary_df.loc[:, ["Spend per Student","Average Math Score", "Average Reading Score",
                                     "% Passing Math", "% Passing Reading",
                                     "% Overall Passing"]].groupby("Spend per Student")

school_spend.mean()
```
